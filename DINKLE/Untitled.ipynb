```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
result = pd.read_csv("experiment_result.csv")
```

```python
min_x = result[result["MAE"] == result["MAE"].min()]["epsilon for error"]
min_y = result["MAE"].min()
```

```python
plt.figure(figsize=(10,5))
plt.plot(result["epsilon for error"], result["MAE"])
plt.plot(min_x, min_y,'ro',label="point")
plt.xlabel("Epsilon for error") 
plt.ylabel("MAE") 

plt.savefig('Experiment_for_epsilon_for_error.png',bbox_inches='tight')
```
